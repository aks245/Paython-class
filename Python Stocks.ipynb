```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
glaxo=(pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Datasets/main/GLAXO.csv"))[['Date','Close']]
beml=(pd.read_csv("https://raw.githubusercontent.com/aishwaryamate/Datasets/main/BEML.csv"))[['Date','Close']]
```

```python
print (glaxo)
print (beml)
```

```python
#Data Description
```

```python
print (glaxo.describe()) 
print (beml.describe())
```

```python
print (glaxo.info())
print (beml.info())
```

```python
#Change the data type of date column to Datetime
```

```python
glaxo['Date']=pd.DatetimeIndex(glaxo['Date'])
beml['Date']=pd.DatetimeIndex(beml['Date'])
#The data type of date column is changed to datetime from object
```

```python
#setting date column as a index column
glaxo=glaxo.set_index(glaxo['Date'])
beml=beml.set_index(beml['Date'])
```

```python
print (glaxo)
print (beml)
```

```python
#Deleting original Date column
glaxo.drop(columns=['Date'], inplace=True)
beml.drop(columns=['Date'], inplace=True)
```

```python
print (glaxo)
print (beml)
```

```python
#To check the data tpes of columns
#we have changed Date column to datetime formate and then to index
print (glaxo.columns)
print (beml.columns)
```

```python
print (glaxo)
print(beml)
```

```python
#plot closing value for both glaxo and beml
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 200
plt.plot(glaxo['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
plt.plot(beml['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
#calculate gain
glaxo['Gain']=glaxo['Close'].pct_change(periods=1)
beml['Gain']=beml['Close'].pct_change(periods=1)
print (glaxo)
print (beml)
```

```python
glaxo.dropna(inplace=True)
beml.dropna(inplace=True)
print (glaxo)
print (beml)
```

```python
#plot Gain 
plt.plot(glaxo['Gain'])
plt.axhline(glaxo['Gain'].mean(), color='r')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.plot(beml['Gain'])
plt.axhline(beml['Gain'].mean(), color='r')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
#subplot for distribution of BEML and Glaxo
plt.subplot(1,2,1)
sns.distplot(glaxo['Gain'],color = 'r', label='Glaxo')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.title('Distribution Graph')

plt.subplot(1,2,2)
sns.distplot(beml['Gain'],color = 'b', label='BEML')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.title('Distribution Graph')
```

```python
#Calculate mean value
round(glaxo['Gain'].mean(),4)
```

```python
round(beml['Gain'].mean(),4)
```

```python
print (glaxo['Gain'].std())
print (beml['Gain'].std())
```

```python
#Probability of making 2% loss or higher in Glaxo
from scipy import stats
```

```python
#stats.norm.cdf(point,mean.sd)
stats.norm.cdf(-0.02,round(glaxo['Gain'].mean(),4),glaxo['Gain'].std())
```

```python
#probability of fmaking 2% gain or higher in glaxo
1 - stats.norm.cdf(0.02,round(glaxo['Gain'].mean(),4),glaxo['Gain'].std())
```

```python
#Probability of making 2% loss or higher in Beml
stats.norm.cdf(-0.02,round(beml['Gain'].mean(),4),beml['Gain'].std())
```
